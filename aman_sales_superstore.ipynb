```python
market["Date"]=pd.to_datetime(market["Date"])  #converting datatype of column "Date"
market["Date"]
```

```python
market["Time"]=pd.to_datetime(market["Time"])  #converting datatype of column "Time"
market.info()
```

```python
market["year"]=market["Date"].dt.year      #separating "year" from "Date" column.
market["month"]=market["Date"].dt.month    #separating "month" from "Date" column.
market["day"]=market["Date"].dt.day        #separating "day" from "Date" column.
market["hour"]=market["Time"].dt.hour      #separating "hour" from "Time" column.
market.head(5)
```

```python
sns.histplot(data=market['Total_sales'],kde=True)                           # checking the skewness of the column "total_sales"
plt.axvline(x=market.Total_sales.mean(),color='red',alpha=0.5,label='mean')
plt.axvline(x=market.Total_sales.median(),color='blue',ls="--",alpha=0.5,label='median')
plt.legend()
```
